```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = 'Moses supposes his toeses are roses but moses supposes erroneously'
doc = nlp(text)

for token in doc:
    print(f'word : ({token.text}), type : ({token.tag_}) , means : ({spacy.explain(token.tag_)})')
```
